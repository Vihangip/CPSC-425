```python
from PIL import Image
import numpy as np
```

```python
im = Image.open('peacock.png')
```

```python
print(im.size, im.mode, im.format)
```

```python
import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
```

```python
im = im.convert('L')
```

```python
plt.imshow(im)
plt.show()
```

```python
im2 = im.crop((475,130,575,230))
```

```python
im2.save('peacock_head.png','PNG')
```

```python
pea_head = Image.open('peacock_head.png')
```

```python
plt.imshow(pea_head)
plt.show()
```

```python
im2_array = np.asarray(im2)
print(im2_array)
```

```python
average = np.mean(im2_array)
print(average)
```

```python
im3_array = im2_array.copy()
```

```python
for x in range(0,100):
            for y in range(0,100):
                im3_array[y,x] = min(im3_array[y,x] + 50, 255)
```

```python
im3 = Image.fromarray(im3_array)
```

```python
im3.save('peacock_head_bright.png','PNG')
```

```python
pea_head_bright = Image.open('peacock_head_bright.png')
plt.imshow(pea_head_bright)
plt.show()
```

```python
im4_array = im2_array.copy()
```

```python
im4_array = im4_array * 0.5
```

```python
im4_array = im4_array.astype('uint8')
```

```python
im4 = Image.fromarray(im4_array)
```

```python
im4.save('peacock_head_dark.png','PNG')
```

```python
pea_head_dark = Image.open('peacock_head_dark.png')
plt.imshow(pea_head_dark)
plt.show()
```
